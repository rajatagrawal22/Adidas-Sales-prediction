```python
encode_col=['Retailer', 'Retailer ID','Region', 'State', 'City','Product','Sales Method']
df_adidas[encode_col] = df_adidas[encode_col].apply(LabelEncoder().fit_transform)
df_adidas.head()
```

```python
# import datetime
# split_date = datetime.date(2021,6,31)
# df_train=df[(pd.to_datetime(df_adidas[df_adidas.columns[0]]) < split_date)]
# df_train.tail()
```

```python
df_adidas['Invoice_date'] = [d.date() for d in df_adidas['Invoice Date']]
df_adidas['Invoice_time'] = [d.time() for d in df_adidas['Invoice Date']]
```

```python
import datetime
split_date = datetime.date(2021,6,30)
# df_train=df_adidas[(pd.to_datetime(df_adidas[df_adidas['Invoice_date']]) < split_date)]
# df_train.tail()\
df_train=df_adidas[df_adidas['Invoice_date'] < split_date]
df_test=df_adidas[df_adidas['Invoice_date'] > split_date]
```

```python
df_train.shape,df_test.shape
```

```python
X= df_adidas[['Retailer', 'Retailer ID', 'Region', 'State', 'City',
       'Product', 'Price per Unit', 'Units Sold', 'Operating Profit', 'Operating Margin', 'Sales Method']]
Y=df_adidas['Total Sales']
```

```python
df_adidas["Region"].value_counts()
```

```python
for i in X.columns:
    X[i]=X[i].astype('int')
```

```python
X.dtypes
```

```python
import xgboost
model = xgboost.XGBRegressor()
# model = xgboost.XGBRegressor(n_estimators=1000, max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8)
# model = xgboost.XGBRegressor()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
from math import sqrt
rms = sqrt(mean_squared_error(y_test, y_pred))
rms
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=5, random_state=0)
# regr = RandomForestRegressor(random_state=43)
regr.fit(x_train, y_train)
```

```python
regr.predict(x_test)
y_pred_reg = regr.predict(x_test)
y_pred_reg
```

```python
rms = sqrt(mean_squared_error(y_test, y_pred_reg))
rms
```

```python
r2_score(y_test, y_pred_reg)
```
